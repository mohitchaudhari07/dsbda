```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("heart.csv")
df.head()
```

```python
# Set style
sns.set(style="whitegrid")
```

```python
# 1. Countplot of Heart Disease cases
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='target', palette='Set2')
plt.title("Heart Disease Presence (1 = Yes, 0 = No)")
plt.show()
```

```python
# 2. Distribution of Age
plt.figure(figsize=(8, 4))
sns.histplot(df['age'], bins=20, kde=True, color='blue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# 3. Chest Pain Type vs Target
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x=df['cp'].astype(str), hue=df['target'].astype(str), palette='Set1')
plt.title("Chest Pain Type vs Heart Disease")
plt.xlabel("Chest Pain Type")
plt.ylabel("Count")
plt.legend(title="Heart Disease (1=Yes, 0=No)")
plt.show()
```

```python
# 4. Heatmap of Correlation Matrix
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title("Feature Correlation with Heatmap")
plt.show()
```
